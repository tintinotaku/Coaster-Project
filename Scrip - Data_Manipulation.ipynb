```python
df = df[['coaster_name',
#    'Length', 'Speed', 
   'Location', 'Status', 
# 'Opening date',
      #  'Type', 
       'Manufacturer', 
    #    'Height restriction', 'Model', 'Height',
    #    'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
    #    'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
    #    'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
    #    'Track layout', 'Fastrack available', 'Soft opening date.1',
    #    'Closing date', 'Opened', 'Replaced by', 'Website',
    #    'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
    #    'Single rider line available', 'Restraint Style',
    #    'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 
       'latitude', 'longitude', 
       'Type_Main',
       'opening_date_clean', 
    #    'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
      #  'height_value', 
    #    'height_unit', 
    'height_ft',
       'Inversions_clean', 'Gforce_clean']]
df
```

```python
df_use = df.copy()
df_use.head()
```

```python
df_use.dtypes
```

```python
#Rename the column

df_use.rename(columns ={'Inversions_clean':"Inversion",
                        'coaster_name':"Coaster",
                        'year_introduced':"Year_introduced",
                        'opening_date_clean':"Opening Date",
                        'speed_mph':"Speed",
                        'Type_Main':"Type",
                        'height_ft':"Height",
                        'Gforce_clean':"Gforce",}, inplace = True)
df_use.head()
```

```python
#Checking missing data
#Or can use this function: df.isna().sum()

for col in df_use.columns:
    a = np.mean(df_use[col].isnull())
    print('{} - {}%'.format(col, a*100))
    
```

```python
#Finding the duplicate data
df_use.loc[df_use.duplicated(subset=['Coaster'])]
```

```python
#Look at duplicate data
df_use.query('Coaster == "Crystal Beach Cyclone"')
```

```python
#This "~" mean you will take the inverse data of this function: df_use.loc[df_use.duplicated(subset=['Coaster', 'Location', 'Opening Date'])].reset_index(drop=True)
#What makes us remove all 97 duplicated data above

df = df_use.loc[~df_use.duplicated(subset=['Coaster', 'Location', 'Opening Date'])].reset_index(drop=True).copy()

df
```

```python
# Top 10 years coasters introduced by using bar chart

df['Year_introduced'].value_counts().nlargest(10).plot(kind = 'bar', xlabel = 'Year', ylabel = 'Total Coaster', title = 'Top 10 Years Coasters Introduced',color = sns.color_palette('viridis'))
```

```python
#Histogram chart for Speed(mph)

df['Speed'].plot(kind = 'hist', bins =20, xlabel = 'Speed(mph)')
```

```python
df['Speed'].plot(kind ='kde')
```

```python
#Scatter plot between Speed and Height
#You can use this function: ------ df.plot(kind = 'scatter', x='Speed', y='Height')

sns.regplot(x='Speed', y='Height', data =df, line_kws = {'color':'red'})
plt.title('Scatter chart between Speed & Height', fontsize = 15)
```

```python
sns.scatterplot(x='Speed', y='Height', hue = 'Year_introduced', data =df)
plt.title('Scatter chart between Speed & Height by year', fontsize = 15)
```

```python
sns.pairplot(data =df, vars = ['Speed', 'Height', 'Year_introduced', 'Inversion', 'Gforce'], hue = 'Type')
```

```python
corr = df[['Speed', 'Height', 'Year_introduced', 'Inversion', 'Gforce']].dropna().corr(method ='pearson')
corr
```

```python
a = sns.heatmap(data = corr, cmap = 'magma', annot= True)
a.set_title('Correlation Matrix', fontsize = 15)
a.set(xlabel= 'Features', ylabel= 'Features')
```

```python
#What are the locations with the fastest roller coasters

df.query('Location != "Other"').groupby('Location')['Speed']\
    .agg(['mean', 'count'])\
    .query('count >=10')\
    .sort_values('mean')['mean']\
    .plot(kind= 'barh', title = 'Average Coast Speed by Location ', figsize = (12,8), xlabel = 'Average Speed(mph)')
```
